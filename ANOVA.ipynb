```python
# import libraries
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols   # ordinary least square
```

```python
# creating the dataframe as wide format
data = {'Toddlers': [67, 42, 67, 56, 62, 64, 59, 72, 71, 60],
        'Teens': [50, 52, 43, 67, 67, 59, 67, 64, 63, 65],
        'Adults': [48, 49, 50, 55, 56, 61, 61, 60, 59, 64],
        'SenCitz': [47, 67, 54, 67, 68, 65, 65, 56, 60, 65]}

df = pd.DataFrame(data)
df
```

```python
# Convert wide format data to long format data. This process is called as data melt

data_melt = df.melt(var_name='Groups', value_name='Values')
data_melt
```

```python
# hypothesis
    # Ho : All age groups has same engagemnet
    # Ha : Not all groups has same engagemnet. Aleast one is different

# interpretation
# we accept Null Hypothesis (Ho) if,
    # F value is close to 1
    # P value > 0.05 (alpha)

# OLS
model = ols('Values ~ C(Groups)', data=data_melt).fit()

# Perform ANOVA
anova_table = sm.stats.anova_lm(model)      # anova_lm:  anova_linear-model

# Print the output
print(anova_table)
```

```python
# R-Squared
r_square = model.rsquared
print('R-Square:', round(r_square*100, 2), '%')
```
